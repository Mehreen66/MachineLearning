```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter

#SG Filter: Applies a Savitzky-Golay filter to smooth the IMFs and residue, removing noise and enhancing the signal
#for better prediction accuracy.


# Example data
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x) + np.random.normal(0, 0.1, 100)  # Noisy sine wave

# Apply Savitzky-Golay filter
window_length = 11  # Must be odd
polyorder = 2  # Polynomial order

y_filtered = savgol_filter(y, window_length, polyorder)

# Plot the original and filtered data
plt.plot(x, y, label='Noisy Signal')
plt.plot(x, y_filtered, color='red', label='Filtered Signal')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Savitzky-Golay Filter')
plt.show()
```
