```python
import numpy as np
import matplotlib.pyplot as plt
from PyEMD import EMD


#EMD decomposes a signal into Intrinsic Mode Functions (IMFs) without any added noise. However, it may suffer from mode mixingn
#  (signal processing, mode mixing makes it hard to distinguish between different components of a signal, like separating different
#   musical instruments in a recording..)

# Generate a sample dataset (e.g., a noisy sine wave)
t = np.linspace(0, 1, 1000)
s = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(1000)

# Initialize EMD
emd = EMD()

# Apply EMD to the dataset
imfs = emd(s)

# Plot the original signal and its IMFs
plt.figure(figsize=(12, 9))

plt.subplot(len(imfs) + 1, 1, 1)
plt.plot(t, s, 'r')
plt.title("Original Signal")

for i, imf in enumerate(imfs):
    plt.subplot(len(imfs) + 1, 1, i + 2)
    plt.plot(t, imf, 'g')
    plt.title(f"IMF {i + 1}")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from PyEMD import CEEMDAN

#CEEMDAN reduces mode mixing by adding white Gaussian noise to the signal and decomposing multiple realizations.

# Generate a sample dataset (e.g., a noisy sine wave)
t = np.linspace(0, 1, 1000)
s = np.sin(2 * np.pi * 5 * t) + 0.5 * np.random.randn(1000)

# Initialize CEEMDAN
ceemdan = CEEMDAN()

# Apply CEEMDAN to the dataset
imfs = ceemdan(s)

# Plot the original signal and its IMFs
plt.figure(figsize=(12, 9))

plt.subplot(len(imfs) + 1, 1, 1)
plt.plot(t, s, 'r')
plt.title("Original Signal")

for i, imf in enumerate(imfs):
    plt.subplot(len(imfs) + 1, 1, i + 2)
    plt.plot(t, imf, 'g')
    plt.title(f"IMF {i + 1}")

plt.tight_layout()
plt.show()
```
